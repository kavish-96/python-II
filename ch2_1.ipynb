```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_node(1)
G.add_nodes_from([2,3,4])
G.add_nodes_from(range(5,8))
G.add_edge(1,2)
G.add_edges_from([(2,3),(4,5),(1,6),(7,5),(3,1),(5,6),(3,2)])
nx.draw_networkx(G,node_size=700,node_color='red',edge_color='blue',width=4)
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
L = [[1,0,0],[0,1,1],[1,1,1]]
G.add_nodes_from([0,1,2])

for i in range(3):
    for j in range(3):
        if L[i][j]==1:
            G.add_edge(i,j)
            
nx.draw_networkx(G,node_size=700,node_color='red',edge_color='blue',width=4)
plt.show()
```

```python
# QB-72
# The following dictionary shows how five people follow each other on Instagram:
# instagram = {'person1': [0,1,1,0,1], 'person2': [0,0,1,0,1], 'person3': [1,1,0,1,1],'person4': [1,1,1,0,0], 'person5': [1,1,0,0,0]}
# E.g., the list for person1 has the value on index 2 as 1 which means person1 followsperson3 and a directed edge should be 
# added from person1 to person3.
# Using networkx library, create a directed graph.
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
instagram = {'person1': [0,1,1,0,1], 'person2': [0,0,1,0,1], 'person3': [1,1,0,1,1],'person4': [1,1,1,0,0], 'person5': [1,1,0,0,0]}
x = list(instagram.values())

G.add_nodes_from([0,1,2,3,4])

for i in range(5):
    for j in range(5):
        if x[i][j]==1:
            G.add_edge(i,j)
            
nx.draw_networkx(G,node_size=400,node_color='red',edge_color='blue',width=3)
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

months = np.arange(12)
income = np.array([5,9,6,6,10,7,6,4,4,5,6,4]) # in thousands
expense = np.array([6,6,8,3,6,9,7,8,6,6,4,8]) # in thousands

fig,ax = plt.subplots(figsize=(8,8))
ax.plot(months,income,color='green')
ax.plot(months,expense,color='red')
ax.fill_between(months,income,expense,where=(income>expense),color='green',alpha=0.3,label='positive',interpolate=True)
ax.fill_between(months,income,expense,where=(income<expense),color='red',alpha=0.3,label='negative',interpolate=True)
plt.ylabel('income/expense')
plt.legend()
plt.show()
```
