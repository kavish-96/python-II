```python
from pywaffle import Waffle
import pandas as pd
import matplotlib.pyplot as plt

data = {'phone':['Samsung','mi','1+','apple','oppo'],'stock':[15,26,7,18,25]}
df = pd.DataFrame(data)
plt.figure(FigureClass=Waffle, rows=8, values=df.stock, labels=list(df.phone))
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = np.random.randint(low=1,high=100,size=(10,10))
print(data)
sns.heatmap(data=data, vmin=30, vmax=70, cmap='tab20',annot=True, linewidths=2, linecolor='black', cbar=True, xticklabels=True, yticklabels=True)
plt.show()
```
