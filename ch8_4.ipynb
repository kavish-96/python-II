```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np

html_file = "Certified used Mercedes-Benz for sale.html"
file = open(html_file,encoding="utf-8")
response = file.read()

soup = BeautifulSoup(response , "html.parser")
# print(soup.prettify)

all_cars = soup.find_all("div",class_="vehicle-details")
name = []
rate = []
reviews = []
dealer = []
price = []
mi = []
for i in all_cars:
    name.append(i.find("a",class_="vehicle-card-link").text.strip())
    
    mi.append(i.find("div",class_="mileage").text.strip())
    
    price.append(i.find("span",class_="primary-price").text.strip())
    
    try:
        rate.append(i.find("span",class_="sds-rating__count").text.strip())
    except:
        rate.append(np.nan)
        
    reviews.append(i.find("span", class_= "sds-rating__link sds-button-link").text.strip())
    
    dealer.append(i.find("div", class_="dealer-name").text.strip())
    
d = {"car":name, "dealer":dealer, "price":price, "mileage":mi, "ratings":rate, "review_count":reviews}
df = pd.DataFrame(d)
df
```

```python
## indeed jobs html file scraping (code is correct, error due to file explorer problem)

import requests
from bs4 import BeautifulSoup
import numpy as np
html_file = "indeed jobs.html"
file = open(html_file,encoding="utf-8")
response = file.read()
soup = BeautifulSoup(response,"html.parser")
print(soup.prettify)
```

```python
import numpy as np
all_resume = soup.find_all("li",class_="css-1ac2h1w eu4oa1w0")[0:-1]
name_of_job = []
name_of_company = []
job_location = []
job_details = []
for i in all_resume :
    try:
        name_of_job.append(i.find("a",class_="jcs-JobTitle css-1baag51 eu4oa1w0").text.strip())
        name_of_company.append(i.find("span",class_="css-1h7lukg eu4oa1w0").text.strip())
        job_location.append(i.find("div",class_="css-1restlb eu4oa1w0").text.strip())
        job_details.append(i.find("div",class_="jobMetaDataGroup css-qspwa8 eu4oa1w0").text.strip())
    except:
        name_of_job.append(np.nan)
        name_of_company.append(np.nan)
        job_details.append(np.nan)
        job_location.append(np.nan)
           
import pandas as pd
d = {"name-of-job":name_of_job,"name-of-company":name_of_company,"job-location":job_location,"job-details":job_details}
df = pd.DataFrame(d)
df
```

```python
## flipkart.html file :

import requests
from bs4 import BeautifulSoup
import numpy as np
html_file = "flipkart.html"
file = open(html_file,encoding="utf-8")
response = file.read()
soup = BeautifulSoup(response,"html.parser")
print(soup.prettify)
```

```python
import numpy as np
all_phone = soup.find_all("div",class_="_2kHMtA")
phone_name = []
ratings = []
flipkart_price = []
market_price = []
for i in all_phone:
    phone_name.append(i.find("div",class_="_4rR01T").text.strip())
    ratings.append(i.find("div",class_="_3LWZlK").text.strip())
    flipkart_price.append(i.find("div",class_="_30jeq3 _1_WHN1").text.strip())
    try:
        market_price.append(i.find("div",class_="_3I9_wc _27UcVY").text.strip())
    except:
        market_price.append(np.nan)
# print(ratings)        
import pandas as pd
d = {"Phone_name":phone_name,"Ratings":ratings,"Flipkart_Price":flipkart_price,"Market_price":market_price}
df = pd.DataFrame(d)
df
```
