```python
# read csv file ipl-matches.csv, create data frame then find :
# 1. how many matches gone in super over
# 2. how many matches won by toss winner
# 3. how many matches won by chennai super kings at kolkata
# 4. In how many matches, hardik pandya is player of match against rajasthan royals

import pandas as pd

df = pd.read_csv('ipl-matches.csv')
# print(df.to_string())

print(len(df[df['SuperOver']=='Y'])) #1
print(len(df[df['TossWinner']==df['WinningTeam']])) #2
print(len(df[(df['City']=='Kolkata') & (df['WinningTeam']=='Chennai Super Kings')])) #3
print(len(df[(df['Player_of_Match']=='HH Pandya') & ((df['Team1']=='Rajasthan Royals') | (df['Team2']=='Rajasthan Royals'))]))
```

```python
# QB-34

# Read this csv file car data.csv and display the basic information like memory and data types for this data frame.
#  Write Python Code for the following cases:
#  1.How Many ritz car are there with kms driven more than 30000km.
#  2.How many Petrol cars are of the year 2017 and whose selling price > 10 lakhs. 
#  3.How many swift cars are there with selling price < 4 Lakhs.
#  4.How many Automatic Transmission Petrol Car are of the year 2015 whose selling price is > 10 Lakhs.
#  5.List out Vehicles with Automatic Transmission and selling price < 1 Lakh.
#  6.How Many Petrol Vehicles are there with kms driven more than 50000kms and Year is between 2010 to 2015(both Year 
# included).
#  7.List out the cars whose price difference between present price and selling price is > 15 lakhs.
#  8.How many Petrol vehicles are there whose kms driven < 5000km and selling price < 50000.

import pandas as pd

df = pd.read_csv('car data.csv')
print(len(df[(df['Car_Name']=='ritz') & (df['Kms_Driven']>30000)])) #1
print(len(df[(df['Fuel_Type']=='Petrol') & (df['Year']==2017) & (df['Selling_Price']>10)])) #2
print(len(df[(df['Car_Name']=='swift') & (df['Selling_Price']<4)])) #3
print(len(df[(df['Fuel_Type']=='Petrol') & (df['Transmission']=='Automatic') & (df['Year']==2015) & (df['Selling_Price']>10)])) #4
print(len(df[((df['Transmission']=='Automatic') & (df['Selling_Price']<1))])) #5
print(len(df[(df['Fuel_Type']=='Petrol') & (df['Kms_Driven']>50000) & ((df['Year']>=2010) & (df['Year']<=2015))])) #6
print(len(df[(((df['Present_Price']-df['Selling_Price'])>15))])) #7
print(len(df[(df['Fuel_Type']=='Petrol') & (df['Kms_Driven']<5000) & (df['Selling_Price']<0.5)])) #7
```
