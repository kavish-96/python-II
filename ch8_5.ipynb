```python
## SENSEX Top Gainers, BSE Daily Top Gainers - Stock Market.html file

import requests
from bs4 import BeautifulSoup
import numpy as np

html_file = "SENSEX Top Gainers, BSE Daily Top Gainers - Stock Market.html"
file = open(html_file,encoding="utf-8")
response = file.read()
soup = BeautifulSoup(response,"html.parser")
print(soup.prettify)
```

```python
import numpy as np
all_details = soup.find_all("table",class_="common-table")

company = []
currentprice = []
prevclose = []
date = []
change = []
changepr = []
dayslow = []
dayshigh = []

for i in all_details :
    
    x = i.find_all("tr")
    
    for j in x[1:] :
        
        a = j.text.split("\n")
        b=[]
        for k in a :
            if k=="" :
                continue
            b.append(k)
        
        company.append(b[0])
        currentprice.append(b[1])
        prevclose.append(b[2])
        date.append(b[3])
        change.append(b[4])
        changepr.append(b[5])
        dayslow.append(b[6])
        dayshigh.append(b[7])
        
d = {"company":company, "current price":currentprice, "prev close":prevclose, "date":date, "change":change, "change pr":changepr, "days low":dayslow, "days high":dayshigh}

import pandas as pd

df = pd.DataFrame(d)
df
```
