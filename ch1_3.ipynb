```python
## outliers

import pandas as pd
import numpy as np

df1 = pd.read_csv('auto-mpg.csv')

Q1 = df1['acceleration'].quantile(0.25)
Q3 = df1['acceleration'].quantile(0.75)
IQR = Q3-Q1 # inter-quantile rank
high_val = Q3 + (1.5 * IQR)
low_val = Q1 - (1.5 * IQR)
print(high_val,low_val)

# df_outliers = df1.loc[(df1['acceleration']<low_val)|(df1['acceleration']>high_val)]
# print(df_outliers) # prints outlying rows 

df_removed = df1.loc[(df1['acceleration']>=low_val)&(df1['acceleration']<=high_val)]
print(df_removed) # prints non-outlying rows
```
