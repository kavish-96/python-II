```python
### Regular Expression

import re

# findall()

txt = 'The rain in Surat'
x = re.findall('ra',txt)
print(x) # ['ra', 'ra']

x = re.findall('[rai]',txt)
print(x) # ['r', 'a', 'i', 'i', 'r', 'a']
print(len(x)) # 6

x = re.findall("xyz",txt)
print(x) # []
```

```python
# search()

txt = 'The rain in Surat'

x = re.search('\s',txt)
print(x) # <re.Match object; span=(3, 4), match=' '>
print(x.start()) # 3

x = re.search("xyz",txt)
print(x) # None

```

```python
# split()

x=re.split("\s",txt)
print(x) # ['The', 'rain', 'in', 'Surat']

x=re.split("\s",txt,2)
print(x) # ['The', 'rain', 'in Surat']
```

```python
# sub()

x = re.sub('\s','9',txt)
print(x) # The9rain9in9Surat

x = re.sub('\s','9',txt,2)
print(x) # The9rain9in Surat
```

```python
import re
print(re.split('\s+','chrome is better than explorer',maxsplit=3)) # ['chrome', 'is', 'better', 'than explorer']

t = 'is this python?'
p = '\w{2}\W+[^\W]'
r = re.sub(p,'**',t)
print(r) # **h**ython?

print('હેલ્લો વોર્લ્ડ')
```

```python
t = 'the code is AAA333BB and PQR365RRR'
p = '[A-Z]{3}\d{3}[A-Z]{3}'
r = re.findall(p,t)
print(r) # ['PQR365RRR']
print(r[0][0])

txt = '08 times before 11:45 AM'
x = re.findall('[1-5][0-9]',txt)
print(x) # ['11', '45']
```

```python
t = 'Hello, how are you?'
p = '\w{3}\W+'
r = re.sub(p,'###',t)
print(r) # He############

t = 'quick brown fox jumps over the lazy dog.'
r = re.findall('\w{4}\s',t)
print(r) # ['uick ', 'rown ', 'umps ', 'over ', 'lazy ']

p = 'bl\w+\W'
s = 'black, blue and brown'
m = re.findall(p,s)
print(m) # ['black,', 'blue ']
```
